```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/sparcs-2022.csv')
```

```python
# extracting relevant columns
df1 = df[['Age Group','Gender','Length of Stay','Total Charges','Total Costs', 'Type of Admission']]
```

```python
df1
```

```python
## basic descriptive statistics ##
```

```python
# length of stay
# converting length of stay to numeric first
df1['Length of Stay'] = pd.to_numeric(df1['Length of Stay'], errors='coerce')
df1['Length of Stay'].describe()
```

```python
# total charges
# converting total charges to numeric first
df1['Total Charges'] = pd.to_numeric(df1['Total Charges'], errors='coerce')
df1['Total Charges'].describe()
```

```python
# total costs
# converting total costs to numeric first
df1['Total Costs'] = pd.to_numeric(df1['Total Costs'], errors='coerce')
df1['Total Costs'].describe()
```

```python
## categorical values ##
```

```python
# age group
df1['Age Group'].value_counts()
```

```python
age_group_counts = df1['Age Group'].value_counts()
age_group_counts.plot(kind='bar', color='skyblue')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.title('Distribution of Age Groups')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```

```python
# gender
df1['Gender'].value_counts()
```

```python
gender_counts = df1['Gender'].value_counts()
gender_counts.plot(kind='bar', color='skyblue')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Distribution of Genders')
plt.xticks(rotation=0, ha='center')
plt.tight_layout()
plt.show()
```

```python
# type of admission
df1['Type of Admission'].value_counts()
```

```python
admission_counts = df1['Type of Admission'].value_counts()
admission_counts.plot(kind='bar', color='skyblue')
plt.xlabel('Type of Admission')
plt.ylabel('Count')
plt.title('Distribution of Admission Types')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability if necessary
plt.tight_layout()
plt.show()
```

```python
## visualizations ##
```

```python
# histogram for length of stay
sns.histplot(data=df1, x='Length of Stay')
plt.title('Distribution of Length of Stay')
plt.xlabel('Length of Stay')
plt.ylabel('Frequency')
plt.show()
```

```python
# boxplot for total charges
sns.boxplot(data=df1, y='Total Charges')
plt.title('Boxplot of Total Charges')
plt.ylabel('Total Charges')
plt.show()
```

```python
# bar plot for type of admission
sns.countplot(data=df1, x='Type of Admission')
plt.title('Distribution of Admission Types')
plt.xlabel('Type of Admission')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

```python
age_group_costs = df1.groupby('Age Group')['Total Costs'].mean()
print(age_group_costs)
```

```python
age_group_costs.plot(kind='bar', color='skyblue')
plt.xlabel('Age Group')
plt.ylabel('Average Total Cost')
plt.title('Average Total Cost by Age Group')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
